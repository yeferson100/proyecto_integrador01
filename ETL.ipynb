```python
import pandas as pd
import numpy as np
```

##PROYECTO INDIVIDUAL
* Iniciamos por cargar los datasets

- credits.cvs

- movies_dataset.cvs

- concatenamos ambas DF y eliminamos la columna id que estaria repetida

```python
df_mov = pd.read_csv('movies_dataset.csv')
df_cred = pd.read_csv('credits.csv')
data = pd.concat([df_mov, df_cred], axis=1)
data=data.loc[:, ~data.columns.duplicated()]
```

* Los valores nulos de los campos revenue, budget deben ser rellenados por el número 0.

* Los valores nulos del campo release date deben eliminarse.

* De haber fechas, deberán tener el formato AAAA-mm-dd, además deberán crear la columna release_year donde extraerán el año de la fecha de estreno.

* Crear la columna con el retorno de inversión, llamada return con los campos revenue y budget, dividiendo estas dos últimas revenue / budget cuando no hay datos disponibles para calcularlo, deberá tomar el valor 0.

* Eliminar las columnas que no serán utilizadas, video,imdb_id,adult,original_title,poster_path y homepage.



```python
#Rellenamos valores nulos por Cero de revenue y budget
data.loc[data['revenue'].isnull(), 'revenue']=0
data.loc[data['budget'].isnull(), 'budget']=0
```

```python
#los datos que no sean datetime seran nulos en release_date
data['release_date']=pd.to_datetime(data['release_date'], errors = 'coerce')
data['release_year']= data['release_date'].dt.year
data.isnull().sum()
```

```python
#Los valores nulos del campo release_date son eliminados
data.dropna(subset=['release_date'], inplace=True)
data.isnull().sum()
```

```python
#damos formato a release_date
data['release_date']=data['release_date'].dt.strftime('%Y-%m-%d')
```

* Damos formato numerico a budget, revenue para hacer la división y pasamos de float a int release_year

```python
data[['budget','revenue','release_year']]=data[['budget','revenue','release_year']].astype(int)
data.info()
```

* budget = 0 lo llenamos con NaN para que al dividir por revenue no sea un error matematico

* reemplazamos los valores NaN de return como valor cero

```python
data.loc[data['budget']==0, 'budget']=None
data['return']=data['revenue']/data['budget']
data.loc[data['return'].isnull(), 'return']=0
data
```

* creamos la columna franquicia con la lista de franquicia correspondiente en belongs_to_colletion
* creamos la columna generos con la lista de generos por pelicula
* creamos la columna company con la lista de compañias por pelicula
* creamos la columna pais_prod con la lista de paises donde se llevo acabo la pelicula

```python
import json
import ast

data['franquicia']=data['belongs_to_collection'].apply(lambda x: ast.literal_eval(x)['name'] if isinstance(x , str) else'').tolist()

data['genres']=data['genres'].str.replace("'",'"')
data['generos']=data['genres'].apply(lambda x: [item['name'] for item in json.loads(x)]).tolist()

data['production_companies']=data['production_companies'].apply(ast.literal_eval)
data['company']=data['production_companies'].apply(lambda x: [item['name'] for item in x])

data['production_countries']=data['production_countries'].apply(ast.literal_eval)
data['pais_prod']=data['production_countries'].apply(lambda x: [item['name'] for item in x])

```

* creamos una funcion para iterar sobre la columna crew, ya que necesitamos la relacion Job : Director - Name : Nombre
* lafuncion recorre cada fila de crew y en el dict donde encuentra Job == Director retorna el nombre

```python
def director_extrac(crew):
    crew_list = ast.literal_eval(crew)
    directores = [item['name'] for item in crew_list if item['job']=='Director']
    return directores
data['directors']= data['crew'].apply(director_extrac)
```

* Eliminamos las columnas que no serán utilizadas, video,imdb_id,adult,original_title,poster_path y homepage
* Eliminamos las columnas transformadas belongs_to_collection','production_companies','production_countries

```python
# Eliminacion de columnas transformadas y no significativas
drop_list=['belongs_to_collection','production_companies','production_countries','video','imdb_id','adult','original_title','poster_path','homepage','cast','crew']
data=data.drop(drop_list,axis=1)
data.to_csv('datos.csv', index=False)
```
